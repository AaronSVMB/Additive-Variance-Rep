## For Now (Lazy Code)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate mean 'player.investment' for each 'subsession.round_number' in both DataFrames
baseline_means = baseline_feb27_df.groupby('subsession.round_number')['player.investment'].mean()
random_means = random_feb27_df.groupby('subsession.round_number')['player.investment'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(baseline_means, label='Baseline', marker='o')  # Plot baseline averages with markers
plt.plot(random_means, label='Random', marker='x')  # Plot random averages with markers

plt.title('Average Player Investment by Subsession Round Number')
plt.xlabel('Subsession Round Number')
plt.ylabel('Average Player Investment')
plt.legend()
plt.grid(True)
plt.show()
```
