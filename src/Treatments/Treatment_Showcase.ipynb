## Parametric

```python
from scipy import stats
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
anova_result = stats.f_oneway(treatment_data_df[treatment_data_df['Treatment'] == 'baseline']['player.investment'],
                              treatment_data_df[treatment_data_df['Treatment'] == 'random']['player.investment'],
                              treatment_data_df[treatment_data_df['Treatment'] == 'ambiguous']['player.investment'])

# Check if we should proceed with Tukey's HSD based on ANOVA results
if anova_result.pvalue < 0.05:
    # Significant difference found; proceeding with Tukey's HSD
    tukey = pairwise_tukeyhsd(endog=treatment_data_df['player.investment'], groups=treatment_data_df['Treatment'], alpha=0.05)
    tukey_result = tukey.summary()
else:
    tukey_result = "ANOVA test did not find significant differences, Tukey's HSD not performed."

tukey_result
```

```python
# Calculating mean, standard error (SE), median, and number of observations for each treatment group
group_stats = treatment_data_df.groupby('Treatment')['player.investment'].agg(['mean', 'sem', 'median', 'count'])
group_stats.rename(columns={'mean': 'Mean', 'sem': 'Standard Error', 'median': 'Median', 'count': 'Observations'}, inplace=True)

group_stats
```

```python
# Identifying periods following a shock
treatment_data_df['previous_shock'] = treatment_data_df.groupby(['Treatment', 'participant.code'])['group.realized_adjustment'].shift(1)
treatment_data_df['investment_change'] = treatment_data_df.groupby(['Treatment', 'participant.code'])['player.investment'].diff()

# Identifying instances where investment remained the same following a shock
treatment_data_df['same_investment'] = treatment_data_df['investment_change'] == 0

# Calculating percentages for each treatment and type of shock
result = treatment_data_df.groupby(['Treatment', 'previous_shock'])['same_investment'].mean().unstack() * 100

result
```
